```python
# 신경망 시작한다해야함, 트레인이미지 네트워크에 넣음
from keras import models
from keras import layers
```

```python
network = models.Sequential() # 모델만들겠다, 인풋레이어
# 그 다음 레이어 추가 dense레이어는 인자 3개 필요 dense(히든레이어 갯수(유닛), activation = 활성함수(보통 relu), imput.shape)
network.add(layers.Dense(512, activation = 'relu', input_shape = (28*28,))) # 다음레이어는 input.shape피ㅣㄹ요없음
network.add(layers.Dense(10, activation = 'softmax')) # 출력레이어 0~9, 소프트맥스 10개의 확률ㅈㅓㅁ수가 들어있는 배열
# 손실함수, optimizer, 평가지표(여기선accuracy)
```

```python
# 손실함수, optimizer, 평가지표(여기선accuracy)
network.compile(optimizer = 'rmsprop', 
                loss = 'categorical_crossentropy',
                metrics = ['accuracy'])
```

```python
train_images[2] # 0~9까지로 맞춰줘야함, 28*28로 리쉐이프 네트워크 크기에 맞춰
```

```python
train_images = train_images.reshape((60000, 28*28)) # input레이어 크기와 맞춤
```

```python
test_images = test_images.reshape((10000, 28*28)) # input레이어 크기와 맞춤
```

```python
train_images = train_images.astype('float32')/255
```

```python
test_images = test_images.astype('float32')/255
```

```python
# label을 범주형데잍터로 (카테고리로)
from keras.utils import to_categorical
train_labels = to_categorical(train_labels)
```

```python
train_labels
```

```python
test_labels = to_categorical(test_labels)
```

## 훈련시작

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128) # epochs반복 batchsize 한번에 몇개/ train데이터에 과대적합됨
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels) 
```

```python
test_acc # tarin데이터로 fit많이하면 과대적합
# 과대적합 증후 train>test/ 과소적합 train과test acc별차이없음
```
