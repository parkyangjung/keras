```
file = codecs.open('4BH00002.txt', 'r', encoding = 'utf-16')
file # 코덱파일이라 볼 수 눈 ㅇ봇음
```

```
soup = BeautifulSoup(file, 'html.parser')
```

```
body = soup.select_one('text > body') # 파일의 body만 가져옴
text = body.getText()
```

```
text
```

## 불러온걸 가지 고 단어로 말뭉치를 만들어서 워드투벡터로 벡터화시켜야함

```
# 불러온걸 가지 고 단어로 말뭉치를 만들어서 워드투벡터로 벡터화시켜야함
kkma = Kkma()
```

```
sentence_list = []

for sent in text.split('\n'): # 엔터기준으로 나눔
  sent_kkma_pos = kkma.pos(sent) # 형태소pos
  word_list = []
  for word_pos in sent_kkma_pos:
    word_list.append(word_pos[0])
    
  sentence_list.append(word_list)
```

```
# 말뭉치를 던져주면 gensim에서 알아서 벡터화시켜줌
sentence_list # 각 센텐스마다 스플릿해줘서 말뭉치 민든것 얘를 그대로 넣어주멷노미
model = Word2Vec(sentence_list, min_count= 1) # min_count = 1 한번 출현한 단어들은 벡터라이징
```

```
model.most_similar(positive = '조선') # 조선이랑 가까운 단어가 순서대로 나옴
```

```
model.most_similar(negative = '정치') # 정치와 네거티브
```

```
model.similarity("조선", "한국") # 가까운지
```

```
len(model['조선']) # 모델에서 조선의 벡터의 갯수 뭘쳐도 100나옴
```

```
import pandas as pd
news = pd.read_csv('it_news.csv', names = ['뉴스본문'], sep = ',', header = None, encoding = 'utf-8')
```

```
news[0:1000]['뉴스본문']
```

```
sentence_list = []

for sent in news[0:10000]['뉴스본문']:
  # 명사만 추출
  sent_kkma_pos = kkma.nouns(sent) # 명사뿐 아니라 모든 형태소 를 다 쪼겜 pos // 명사만추출 nouns
  word_list = []
  for word_pos in sent_kkma_pos:
    word_list.append(word_pos)
    
  sentence_list.append(word_list)
print(sentence_list[0:100])
```

```
# 스킵그램 주변단어보고 다음에 뭐나올지
sentence_list # 각 센텐스마다 스플릿해줘서 말뭉치 민든것 얘를 그대로 넣어주멷노미
model = Word2Vec(sentence_list, size = 100) # min_count = 1 한번 출현한 단어들은 벡터라이징 // size 200d 데이터가 많으면 200d도 ㄱㅊ

```

```
model.most_similar('빅데이터') # 조선이랑 가까운 단어가 순서대로 나옴 // 뉴스데이터가 왕조실로보다 자료양이 더 많아서 더 연관된 단어들이 나옴
```

```
model.similarity("구글", "애플") # 가까운지
```

```
model.most_similar(positive = '인공지능') # 정치와 네거티브
```
