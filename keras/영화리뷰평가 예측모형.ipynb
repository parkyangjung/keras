```python
from tqdm import tqdm
```

```python
stemmer = SnowballStemmer('english')

def stemming(phrase):
    words = phrase.split(" ")
    stemmed_words = []
    
    for word in words:
        stemmed_word = stemmer.stem(word)
        stemmed_words.append(stemmed_word)
        
    stemmed_phrase = " ".join(stemmed_words)
    
    return stemmed_phrase

tqdm.pandas(desc = "스태밍중입니다....")

train["Phrase"] = train["Phrase"].progress_apply(stemming)

train.head()
```

```python
tqdm.pandas(desc="스태밍중입니다...(이번에는 test)")

test["Phrase"] = test["Phrase"].progress_apply(stemming)

test.head()
```

```python
def stop_words(phrase):
    phrase = phrase.replace("n't", "not")
    phrase = phrase.replace("disappointments", "disappintment")
    
    return phrase

train['Phrase'] = train["Phrase"].apply(stop_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# vect = CountVectorizer(max_features = 30000, ngram_range=(1,2))

# vect.fit(train["Phrase"])

vect = CountVectorizer(max_features = 30000, ngram_range=(1,2))

vect.fit(train["Phrase"])
```

```python
X_train = vect.transform(train["Phrase"])
print(X_train.shape)
X_train
```

```python
X_test = vect.transform(test["Phrase"])
print(X_test.shape)
X_test
```

```python
y_train = train["Sentiment"]

print(y_train.shape)
y_train.head()
```

```python
train_vector = pd.DataFrame(X_train.toarray(), columns=vect.get_feature_names())
train_vector.head()
```

```python
from sklearn.linear_model import SGDClassifier

model = SGDClassifier(n_jobs=-1)

model
```

```python
from sklearn.model_selection import cross_val_score, GroupKFold

kfold = GroupKFold(n_splits=5)

score = cross_val_score(model, X_train, y_train, cv=kfold, groups=train["SentenceId"]).mean()

print("Score = {}".format(score))
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(predictions.shape)
predictions[0:10]
```
